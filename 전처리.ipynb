# vix 지수 데이터 전처리

```
import pandas as pd
```

```
vix = pd.read_csv("/content/vix.csv", header=None)
vix
```

```
vix['date'] = 0
vix['vix'] = 0

for i in range(len(vix)):
    vix['date'][i] = vix.iloc[i,0].split("\t")[0]
    vix['vix'][i] = float(vix.iloc[i,0].split("\t")[1])

vix=vix[["date","vix"]]
vix['date'] = vix['date'].apply(lambda x:x.split("년 ")[0])+vix['date'].apply(lambda x:x.split("년 ")[1].split("월 ")[0])+vix['date'].apply(lambda x:x.split("년 ")[1].split("월 ")[1].replace("일",""))
vix['date']=pd.to_datetime(vix['date'])

vix 
```

```
vix.to_csv("vix_refined.csv")
```
